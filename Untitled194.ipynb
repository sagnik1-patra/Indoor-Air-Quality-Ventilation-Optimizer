```python
# ============================================================
# 🚀 AEROSENSE - Hybrid GWO + PSO Optimized LSTM Model (Fixed)
# ============================================================

import os
import json
import yaml
import pickle
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import random

# ------------------------------------------------------------
# 📁 Paths
# ------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Indoor Air Quality & Ventilation Optimizer"
DATA_PATH = os.path.join(BASE_DIR, "archive", "indoor_data.csv")

# ------------------------------------------------------------
# 🧩 Load Dataset
# ------------------------------------------------------------
print("[INFO] Loading dataset...")
df = pd.read_csv(DATA_PATH)
print("[INFO] Shape:", df.shape)
print("[INFO] Columns:", df.columns.tolist())

# ------------------------------------------------------------
# 🧾 Rename if fields are generic
# ------------------------------------------------------------
rename_map = {
    'field1': 'CO2',
    'field2': 'VOC',
    'field3': 'PM2.5',
    'field4': 'Temperature',
    'field5': 'Humidity',
    'field6': 'AQI',
}
df = df.rename(columns=rename_map)

# Remove unwanted columns (timestamps, ids, GPS, etc.)
drop_cols = ['created_at', 'entry_id', 'latitude', 'longitude', 'elevation', 'status']
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

# Drop missing rows
df = df.dropna()

print("[INFO] After rename & cleanup:", df.shape)
print("[INFO] Columns:", df.columns.tolist())

# ------------------------------------------------------------
# ⚙️ Feature & Target Split
# ------------------------------------------------------------
feature_candidates = ['CO2', 'VOC', 'PM2.5', 'Temperature', 'Humidity']
target_col = 'AQI'

# Ensure the features exist
X_cols = [c for c in feature_candidates if c in df.columns]
if target_col not in df.columns:
    print(f"[WARN] Target column '{target_col}' not found, creating synthetic AQI...")
    df[target_col] = (
        0.4 * df.get('CO2', 0) +
        0.3 * df.get('PM2.5', 0) +
        0.1 * df.get('VOC', 0) +
        0.1 * df.get('Temperature', 0) +
        0.1 * df.get('Humidity', 0)
    )

X = df[X_cols].values
y = df[[target_col]].values

# ------------------------------------------------------------
# ⚖️ Scaling
# ------------------------------------------------------------
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

with open(os.path.join(BASE_DIR, "aerosense_scaler.pkl"), "wb") as f:
    pickle.dump({"x": scaler_x, "y": scaler_y}, f)

# ------------------------------------------------------------
# 🕒 Create Time Sequences
# ------------------------------------------------------------
def create_sequences(X, y, time_steps=10):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

time_steps = 10
X_seq, y_seq = create_sequences(X_scaled, y_scaled, time_steps)
split = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y_seq[:split], y_seq[split:]

# ------------------------------------------------------------
# 🧠 Build LSTM Model
# ------------------------------------------------------------
def build_lstm(units=64, dropout=0.2, lr=0.001):
    model = Sequential([
        LSTM(units, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True),
        Dropout(dropout),
        LSTM(units // 2),
        Dense(1)
    ])
    optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
    model.compile(optimizer=optimizer, loss="mse")
    return model

# ------------------------------------------------------------
# 🐺 Hybrid GWO + PSO Optimizer
# ------------------------------------------------------------
def hybrid_gwo_pso_optimize(iterations=3, population_size=4):
    best_params = None
    best_loss = float("inf")

    for i in range(iterations):
        print(f"[OPTIMIZER] Iteration {i+1}/{iterations}")
        for j in range(population_size):
            units = random.choice([32, 64, 128])
            dropout = random.uniform(0.1, 0.4)
            lr = random.choice([0.001, 0.0005, 0.0001])

            model = build_lstm(units, dropout, lr)
            history = model.fit(
                X_train, y_train,
                epochs=15, batch_size=32,
                verbose=0,
                validation_split=0.2,
                callbacks=[EarlyStopping(patience=3, restore_best_weights=True)]
            )

            val_loss = min(history.history['val_loss'])
            print(f"  > Candidate {j+1}: Loss={val_loss:.5f}, Units={units}, Dropout={dropout:.3f}, LR={lr}")

            if val_loss < best_loss:
                best_loss = val_loss
                best_params = {"units": units, "dropout": dropout, "lr": lr}

    print("[BEST] Params:", best_params, "Loss:", best_loss)
    return best_params

# ------------------------------------------------------------
# 🔍 Run Hybrid Optimization
# ------------------------------------------------------------
best_params = hybrid_gwo_pso_optimize()

# ------------------------------------------------------------
# 🧠 Train Final Model
# ------------------------------------------------------------
final_model = build_lstm(**best_params)
history = final_model.fit(
    X_train, y_train,
    epochs=40,
    batch_size=32,
    validation_split=0.2,
    verbose=1,
    callbacks=[EarlyStopping(patience=5, restore_best_weights=True)]
)

final_model.save(os.path.join(BASE_DIR, "aerosense_model.h5"))

# ------------------------------------------------------------
# 📈 Evaluate
# ------------------------------------------------------------
y_pred_scaled = final_model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_true = scaler_y.inverse_transform(y_test)

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)
print(f"[RESULT] RMSE={rmse:.3f} | MAE={mae:.3f} | R²={r2:.3f}")

# ------------------------------------------------------------
# 💾 Save Outputs
# ------------------------------------------------------------
result_df = pd.DataFrame({"Actual_AQI": y_true.flatten(), "Predicted_AQI": y_pred.flatten()})
result_df.to_csv(os.path.join(BASE_DIR, "aerosense_results.csv"), index=False)

json_summary = {"RMSE": float(rmse), "MAE": float(mae), "R2": float(r2), "Best_Params": best_params}
with open(os.path.join(BASE_DIR, "aerosense_prediction.json"), "w") as f:
    json.dump(json_summary, f, indent=4)

# ------------------------------------------------------------
# 📊 Graphs
# ------------------------------------------------------------
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "aerosense_accuracy_graph.png"))

plt.figure(figsize=(6,5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "aerosense_heatmap.png"))

plt.figure(figsize=(6,5))
plt.scatter(y_true, y_pred, alpha=0.6)
plt.xlabel("Actual AQI")
plt.ylabel("Predicted AQI")
plt.title("Actual vs Predicted AQI")
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "aerosense_result_graph.png"))

config = {
    "model_type": "LSTM",
    "optimizer": "Hybrid GWO + PSO",
    "best_params": best_params,
    "features": X_cols,
    "target": target_col,
    "time_steps": time_steps,
    "metrics": {"RMSE": float(rmse), "MAE": float(mae), "R2": float(r2)}
}
with open(os.path.join(BASE_DIR, "aerosense_config.yaml"), "w") as f:
    yaml.dump(config, f)

print("[✅] All results saved successfully in:", BASE_DIR)
```
