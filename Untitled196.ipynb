```python
import os, json, yaml, pickle, random
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# ------------------------------------------------------------
# 📁 Paths
# ------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Indoor Air Quality & Ventilation Optimizer"
DATA_PATH = os.path.join(BASE_DIR, "archive", "indoor_data.csv")

# ------------------------------------------------------------
# 🧩 Load Dataset
# ------------------------------------------------------------
print("[INFO] Loading dataset...")
df = pd.read_csv(DATA_PATH)
rename_map = {
    'field1': 'CO2', 'field2': 'VOC', 'field3': 'PM2.5',
    'field4': 'Temperature', 'field5': 'Humidity', 'field6': 'AQI'
}
df = df.rename(columns=rename_map)
drop_cols = ['created_at', 'entry_id', 'latitude', 'longitude', 'elevation', 'status']
df = df.drop(columns=[c for c in drop_cols if c in df.columns])
df = df.dropna()

if 'AQI' not in df.columns:
    df['AQI'] = (0.4*df.get('CO2',0)+0.3*df.get('PM2.5',0)+
                 0.1*df.get('VOC',0)+0.1*df.get('Temperature',0)+0.1*df.get('Humidity',0))

features = [c for c in ['CO2','VOC','PM2.5','Temperature','Humidity'] if c in df.columns]
target = 'AQI'
print(f"[INFO] Using features: {features}")

X, y = df[features].values, df[[target]].values
scaler_x, scaler_y = MinMaxScaler(), MinMaxScaler()
X_scaled, y_scaled = scaler_x.fit_transform(X), scaler_y.fit_transform(y)

with open(os.path.join(BASE_DIR, "aerosense_scaler.pkl"), "wb") as f:
    pickle.dump({"x": scaler_x, "y": scaler_y}, f)

# ------------------------------------------------------------
# ⏳ Sequence builder
# ------------------------------------------------------------
def create_sequences(X, y, time_steps=10):
    Xs, ys = [], []
    for i in range(len(X)-time_steps):
        Xs.append(X[i:i+time_steps]); ys.append(y[i+time_steps])
    return np.array(Xs), np.array(ys)

time_steps = 10
X_seq, y_seq = create_sequences(X_scaled, y_scaled, time_steps)
split = int(0.8*len(X_seq))
X_train, X_test, y_train, y_test = X_seq[:split], X_seq[split:], y_seq[:split], y_seq[split:]

# ------------------------------------------------------------
# 🧠 Build model
# ------------------------------------------------------------
def build_lstm(units=64, dropout=0.2, lr=0.001):
    model = Sequential([
        LSTM(units, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True),
        Dropout(dropout),
        LSTM(units//2),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss="mse")
    return model

# ------------------------------------------------------------
# 🧬 Hybrid AIS + QPSO optimizer
# ------------------------------------------------------------
def hybrid_ais_qpso_optimize(max_iter=4, pop_size=5):
    # Each antibody = [units, dropout, lr]
    def fitness(params):
        u, d, lr = int(params[0]), params[1], params[2]
        model = build_lstm(u, d, lr)
        h = model.fit(X_train, y_train, epochs=15, batch_size=32,
                      verbose=0, validation_split=0.2,
                      callbacks=[EarlyStopping(patience=3, restore_best_weights=True)])
        return min(h.history['val_loss'])
    
    # Initialize population
    pop = np.array([[random.choice([32,64,128]),
                     random.uniform(0.1,0.4),
                     random.choice([0.001,0.0005,0.0001])] for _ in range(pop_size)])
    best_pos, best_score = pop[0], float("inf")
    beta = 1.5  # quantum control

    for iter in range(max_iter):
        print(f"[OPTIMIZER] Iteration {iter+1}/{max_iter}")
        # Evaluate fitness
        scores = np.array([fitness(p) for p in pop])
        idx = np.argmin(scores)
        if scores[idx] < best_score:
            best_score, best_pos = scores[idx], pop[idx].copy()

        # Immune cloning & mutation (AIS)
        clones = []
        for p in pop:
            mutated = p.copy()
            if random.random() < 0.5:
                mutated[0] = random.choice([32,64,128])
            if random.random() < 0.3:
                mutated[1] = np.clip(p[1] + np.random.uniform(-0.05,0.05), 0.05,0.5)
            clones.append(mutated)
        clones = np.array(clones)

        # QPSO update
        mean_pos = np.mean(clones, axis=0)
        for i in range(pop_size):
            fi = random.random()
            u = random.random()
            pop[i] = mean_pos + beta * abs(best_pos - clones[i]) * np.log(1/u) * (1 if fi<0.5 else -1)
            # clamp values
            pop[i][0] = int(np.clip(pop[i][0], 32,128))
            pop[i][1] = float(np.clip(pop[i][1], 0.05,0.5))
            pop[i][2] = float(np.clip(pop[i][2], 0.0001,0.001))
        
        print(f"  Best so far: Loss={best_score:.5f}, Params={best_pos.tolist()}")

    return {"units": int(best_pos[0]), "dropout": float(best_pos[1]), "lr": float(best_pos[2])}

# ------------------------------------------------------------
# 🧩 Run optimizer
# ------------------------------------------------------------
best_params = hybrid_ais_qpso_optimize()
print("[BEST PARAMS]", best_params)

# ------------------------------------------------------------
# 🧠 Train final model
# ------------------------------------------------------------
final_model = build_lstm(**best_params)
hist = final_model.fit(X_train, y_train, epochs=50, batch_size=32,
                       validation_split=0.2, verbose=1,
                       callbacks=[EarlyStopping(patience=5, restore_best_weights=True)])
final_model.save(os.path.join(BASE_DIR, "aerosense_model.h5"))

# ------------------------------------------------------------
# 📊 Evaluate
# ------------------------------------------------------------
y_pred_scaled = final_model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_true = scaler_y.inverse_transform(y_test)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)
print(f"[RESULT] RMSE={rmse:.3f}, MAE={mae:.3f}, R²={r2:.3f}")

# ------------------------------------------------------------
# 💾 Save outputs
# ------------------------------------------------------------
pd.DataFrame({"Actual_AQI":y_true.flatten(),"Predicted_AQI":y_pred.flatten()})\
  .to_csv(os.path.join(BASE_DIR,"aerosense_results.csv"),index=False)
with open(os.path.join(BASE_DIR,"aerosense_prediction.json"),"w") as f:
    json.dump({"RMSE":float(rmse),"MAE":float(mae),"R2":float(r2),"Best_Params":best_params},f,indent=4)

# ------------------------------------------------------------
# 📈 Graphs
# ------------------------------------------------------------
# 1. Accuracy Graph
plt.figure(figsize=(8,5))
plt.plot(hist.history['loss'],label='Train')
plt.plot(hist.history['val_loss'],label='Val')
plt.title("Accuracy Graph – Training vs Validation Loss")
plt.xlabel("Epochs"); plt.ylabel("MSE"); plt.legend()
plt.tight_layout(); plt.savefig(os.path.join(BASE_DIR,"aerosense_accuracy_graph.png"))

# 2. Heatmap
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.tight_layout(); plt.savefig(os.path.join(BASE_DIR,"aerosense_heatmap.png"))

# 3. Result Graph (Scatter)
plt.figure(figsize=(6,5))
plt.scatter(y_true,y_pred,alpha=0.6,color='green')
plt.xlabel("Actual AQI"); plt.ylabel("Predicted AQI")
plt.title("Result Graph – Actual vs Predicted")
plt.tight_layout(); plt.savefig(os.path.join(BASE_DIR,"aerosense_result_graph.png"))

# 4. Prediction Graph (Time-series)
plt.figure(figsize=(10,5))
plt.plot(y_true[:200],label="Actual",color="blue")
plt.plot(y_pred[:200],label="Predicted",color="orange",linestyle="dashed")
plt.legend(); plt.title("Prediction Graph – First 200 Samples")
plt.tight_layout(); plt.savefig(os.path.join(BASE_DIR,"aerosense_prediction_graph.png"))

# 5. Comparison Graph
plt.figure(figsize=(10,5))
sample = 100
plt.plot(y_true[:sample],label="Actual",color='blue')
plt.plot(y_pred[:sample],label="Predicted",color='red',linestyle='--')
plt.fill_between(range(sample),y_true[:sample].flatten(),y_pred[:sample].flatten(),
                 color='gray',alpha=0.2)
plt.title("Comparison Graph – Actual vs Predicted AQI")
plt.xlabel("Sample"); plt.ylabel("AQI"); plt.legend()
plt.tight_layout(); plt.savefig(os.path.join(BASE_DIR,"aerosense_comparison_graph.png"))

# ------------------------------------------------------------
# 🧾 Save YAML Config
# ------------------------------------------------------------
config = {
    "model_type":"LSTM",
    "optimizer":"Hybrid AIS + QPSO",
    "best_params":best_params,
    "features":features,
    "target":target,
    "time_steps":time_steps,
    "metrics":{"RMSE":float(rmse),"MAE":float(mae),"R2":float(r2)}
}
with open(os.path.join(BASE_DIR,"aerosense_config.yaml"),"w") as f:
    yaml.dump(config,f)

print("\n✅ All artifacts generated successfully at:")
print(BASE_DIR)
```
